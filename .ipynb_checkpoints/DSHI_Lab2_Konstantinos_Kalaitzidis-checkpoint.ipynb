## Normalization and Standarisation

Very improtant for K-means clustering. Normalizing the input data before applying the k-Means Clustering algorithm. Normalizing the data is important to ensure that the distance measure accords equal weight to each variable. Without normalization, the variable with the largest scale will dominate the measure ([src](https://www.solver.com/xlminer/help/using-k-means-clustering)). 

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler

min_max_scaler = MinMaxScaler()
X = min_max_scaler.fit_transform(X)

print(f"Minimum value in data is: {np.min(X)}")
print(f"Maximum value in data is: {np.max(X)}")
```

```python
# A plot showing the normalized data
plt.scatter(x=X[:,0], y=X[:,1], c=y, s=30)
plt.xlabel("x1")
plt.ylabel("x2")
plt.grid(True)
plt.title("Synthetic Moon dataset with 100 samples | Real clusters")
plt.show()
```

Function to visualize the dataset and the clusters:

```python
def visualize_clustering_algorithm(X, y, centers=None, title=""):
    """
    Input:
        X: (np.array[N,2]) - The dataset to be clustered
        y: (np.array[N,1]) - Corresponding cluster memberships
        centers: (np.array[K,2]) - [non mandatory parameter]
                    Shows where the current cluster centers are.
    Returns:
        A plot with the dataset and the centroids of the clusters
    """
    plt.scatter(x = X[:,0], y = X[:,1], c=y, s=30, label="Dataset")
    if(centers is not None):
        plt.scatter(x = centers[:,0], y = centers[:,1], c=np.arange(centers.shape[0]), s=200, marker="x", label="Cluster centers")
        plt.legend()
    plt.xlabel("x1")
    plt.ylabel("x2")
    plt.title(title)
    plt.grid(True)
    return plt.show()
```

## Calculating the real cluster centers

```python
def calculate_centers_from_labels(X, labels):
    """
    Input:
        X: (np.array[N,M]) - The dataset to be clustered
        labels: (np.array[N,1]) - Cluster memberships
    Returns:
        centers: (np.array[k,M]) - Coordinates of the cluster centers
    """
    N,M = X.shape
    num_clusters = np.unique(labels).size
    centers = np.zeros((num_clusters,M))
    for i in range(num_clusters):
        data_in_cluster = X[ labels==i ]
        # The centers correspond to the mean value of the data belonging to the cluster i
        centers[i,:] = np.mean(data_in_cluster,axis=0)
    return centers
```

```python
# How does it compare with the real clusters?
real_centers = calculate_centers_from_labels(X, y)
print(f"Based on the labels, there are clusters with centers in: {real_centers}")
```

```python
# Plot the original (real) clusters in the dataset
visualize_clustering_algorithm(X, y, centers=real_centers, title="Real clusters")
```

# Apply K-means from Python library

1. First, we need to initialize an object with the algorithm that we want, in this case it is the `KMeans()` function, which can be customized with multiple parameters ([see docs here](https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html?highlight=kmeans#sklearn.cluster.KMeans)).
2. Then, we need to apply the algorithm in our **training data** by running the function `.fit()` with the corresponding parameters, which creates a **machine learning model**.
3. Lastly, we generate new labels on **unseen test data** by executing the function `.predict()` with the corresponding parameters.
4. A common function is to execute `.fit_predict()` as a shortcut for both actions.

```python
# Importing the necessary functions from sklearn
from sklearn.cluster import KMeans
```

```python
# 1. Initialize a KMeans object that looks for 2 clusters in the data
kmeans = KMeans(n_clusters=2, random_state=RANDOM_SEED)

# 2. Apply the object on our dataset
kmeans.fit(X)

# 3. Extract the estimated labels from the data. 
# (We'd run the function ``.predict()`` only for unseen data, but
# here we just want to actual labels of the original dataset where
# the model was trained)
estimated_clusters_kmeans = kmeans.labels_
print(f"The estimated cluster membership is: {estimated_clusters_kmeans}")
```

```python
# Let's use our function from before to see the result visually
estimated_centers_kmeans = calculate_centers_from_labels(X, estimated_clusters_kmeans)
```

```python
## Simpler visualization of the resulting clusters

def visualize_and_compare_clusters(X, real_labels, real_centers, estimated_labels, estimated_centers):
    """
    Input:
        X: (np.array[N,2]) - The dataset to visualize (only 2 features)
        real_labels: (np.array[N,1]) - Real cluster membership
        real_centers: (np.array[K,2]) - Real cluster centers
        estimated_labels: (np.array[N,1]) - Estimated cluster membership from algorithm
        estimated_labels: (np.array[K,2]) - Estimated cluster centers
    Returns:
        A plot with two axes showing the real clusters and the predicted clusters
    """

    fig, axes = plt.subplots(1, 2, figsize=(10,5))
    
    # First plot contains real cluster membership
    ax = axes[0]
    ax.scatter(x = X[:,0], y = X[:,1], c=real_labels, s=30, label= "Data")
    ax.scatter(x = real_centers[:,0], y = real_centers[:,1], 
            c=np.arange(real_centers.shape[0]), s=200, marker="x", label="Centers")
    ax.set(xlabel="x1",ylabel="x2",title="Real clusters")
    ax.grid(True); ax.legend()

    # Second plot contains estimated cluster membership
    ax = axes[1]
    ax.scatter(x = X[:,0], y = X[:,1], c=estimated_labels, s=30, label="Data")
    ax.scatter(x = estimated_centers[:,0], y = estimated_centers[:,1], 
            c=np.arange(estimated_centers.shape[0]), s=200, marker="x", label="Centers")
    ax.set(xlabel="x1",ylabel="x2",title="Estimated clusters")
    ax.grid(True); ax.legend()

    return plt.show()
```

```python
# Plot the real and the estimated with K-means
visualize_and_compare_clusters(X, 
                               y, 
                               real_centers, 
                               estimated_clusters_kmeans, 
                               estimated_centers_kmeans)
```

## Evaluation Metrics

```python
# Import the codes containing the evaluation metrics
from sklearn import metrics
```

```python
# Internal metrics without knowledge of real values

eval_silhouette_score_kmeans = metrics.silhouette_score(X, estimated_clusters_kmeans, metric="euclidean")
print("The silhouette score for the KMeans Clustering algorithm is:", eval_silhouette_score_kmeans)
```
